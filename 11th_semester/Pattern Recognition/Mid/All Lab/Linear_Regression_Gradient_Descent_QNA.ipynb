```python
# লিনিয়ার রিগ্রেশন কী? গ্রেডিয়েন্ট ডিসেন্ট কীভাবে এখানে কাজ করে?
উত্তর:
লিনিয়ার রিগ্রেশন একটি সুপারভাইজড লার্নিং অ্যালগরিদম যা নির্ভরশীল ও স্বাধীন চলকের মধ্যে রৈখিক সম্পর্ক স্থাপন করে।
গ্রেডিয়েন্ট ডিসেন্ট একটি অপটিমাইজেশন অ্যালগরিদম যা কস্ট ফাংশন মিনিমাইজ করে প্যারামিটার আপডেট করে।

কাজের ধাপ:

1) প্রেডিকশন করা: y_pred = w₁ * X + w₀

2) কস্ট হিসাব করা: cost = (1/(2n)) * Σ(y_pred - y)²

3) গ্রেডিয়েন্ট গণনা: dw₁ = (1/n) * Σ((y_pred - y) * X), dw₀ = (1/n) * Σ(y_pred - y)

4) প্যারামিটার আপডেট: w₁ = w₁ - learning_rate * dw₁, w₀ = w₀ - learning_rate * dw₀

5) কনভার্জ না হওয়া পর্যন্ত ধাপ ১-৪ পুনরাবৃত্তি।
```

```python
# কস্ট ফাংশন (Cost Function) কী? এখানে Mean Squared Error (MSE) ব্যবহার করা হয়েছে কেন?
উত্তর:
কস্ট ফাংশন মাপে মডেলের প্রেডিকশন কতটা ভুল। এখানে ব্যবহৃত হয়েছে MSE:

ব্যবহারের কারণ:

-> সব ভুলকে ধনাত্মক করে (বর্গ করার কারণে)

-> বড় ভুলকে বেশি শাস্তি দেয়

-> গ্রেডিয়েন্ট বের করা সহজ

-> কনভেক্স ফাংশন, তাই গ্লোবাল মিনিমাম নিশ্চিত
```

```python
# লার্নিং রেট (Learning Rate) কী? খুব ছোট বা খুব বড় হলে কী হয়?
উত্তর:
লার্নিং রেট নির্ধারণ করে কত বড় ধাপে প্যারামিটার আপডেট হবে।

খুব ছোট: কনভার্জ হতে অনেক সময় লাগে, লোকাল মিনিমামে আটকে যেতে পারে।

খুব বড়: কস্ট ফাংশন মিনিমাম পেরিয়ে যেতে পারে, কনভার্জ নাও করতে পারে (oscillate করে)।

উপযুক্ত মান: কনভার্জ দ্রুত করে এবং স্থিতিশীল থাকে। (এখানে 0.01 ব্যবহার করা হয়েছে)

```

```python
# ইপোক (Epoch) কী? এখানে 100 ইপোক কেন ব্যবহার করা হয়েছে?
উত্তর:
ইপোক হলো পুরো ট্রেনিং ডেটাসেট একবার ফরোয়ার্ড ও ব্যাকওয়ার্ড পাস করার সংখ্যা।
100 ইপোক ব্যবহারের কারণ:

প্যারামিটার কনভার্জ হতে যথেষ্ট সময় দেয়।

ওভারফিটিং না করেই কম কস্ট নিশ্চিত করে।

কস্ট হিস্টোরি দেখে বোঝা যায় 100 ইপোকের পর কস্ট স্থির হয়ে গেছে।
```

```python
# R-squared (R²) কী? এখানে মান 89.32% এসেছে—এর ব্যাখ্যা কী?
উত্তর:
R² নির্দেশ করে স্বাধীন চলক (X) নির্ভরশীল চলক (y)-এর কতটা ভ্যারিয়েশন ব্যাখ্যা করতে পারে।

এখানে R² = 0.8932 (89.32%) → মানে পিৎজার সাইজ দামের 89.32% ভ্যারিয়েশন
ব্যাখ্যা করতে পারে। বাকি 10.68% অন্যান্য ফ্যাক্টর (টপিং, ব্র্যান্ড, লোকেশন ইত্যাদি)
দ্বারা প্রভাবিত।
```
