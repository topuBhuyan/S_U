```python
# Linear Regression কি?

উত্তর:
Linear Regression হলো এমন একটি পদ্ধতি যেখানে দুইটি ভেরিয়েবলের মধ্যে সরলরৈখিক সম্পর্ক বের করা হয়।
Equation: y = w1*x + w0

w1 = slope
w0 = intercept
```

```python
# এখানে independent ও dependent variable কোনটি?

উত্তর:

Independent variable = size
Dependent variable = price

কারণ size এর উপর price নির্ভর করছে।
```

```python
# Scatter plot কেন ব্যবহার করা হয়েছে?

উত্তর:
Scatter plot ব্যবহার করা হয়েছে size এবং price এর মধ্যে সম্পর্ক visually দেখানোর জন্য।

plt.scatter(df['size'],data['price'])
```

```python
# w0 এবং w1 কি?

উত্তর:

w1 = slope (line এর ঢাল)
w0 = intercept (line y-axis কে যেখানে কাটে)
```

```python
# Overfitting কি?

উত্তর:
Overfitting মানে model training data খুব ভালো শিখে ফেলে কিন্তু নতুন data তে ভালো কাজ করে না।

👉 Training accuracy high
👉 Test accuracy low
```

```python
# slope negative হলে কি বুঝায়?

উত্তর:
Slope negative হলে inverse relationship বোঝায়।

Example:

size বাড়লে price কমে (unusual but possible)
```

```python
# Correlation কি?

উত্তর:
Correlation দুইটি ভেরিয়েবলের সম্পর্কের শক্তি বোঝায়।

-> +1 → strong positive
-> 0 → no relation
-> -1 → strong negative
```

```python
# Residual error কি?

উত্তর:
Actual value ও predicted value এর পার্থক্য।

Residual = Actual - Predicted
```

```python
# Cost function কি?

উত্তর:
Prediction কত ভুল হচ্ছে তা measure করে।

সাধারণত Mean Squared Error ব্যবহার করা হয়।

MSE = Σ (y - y_pred)^2 / n
```

```python
# Regression line best fit মানে কি?

উত্তর:
যে line total error সবচেয়ে কম করে সেটাই best fit line।
```
