```python
import pandas as pd #  প্যান্ডাস লাইব্রেরি ইম্পোর্ট করা হচ্ছে, যা ডেটা টেবিল আকারে দেখতে ও বিশ্লেষণ করতে ব্যবহৃত হয়।
import numpy as np # নিউমেরিক্যাল অপারেশনের জন্য নumpy লাইব্রেরি ইম্পোর্ট।
import matplotlib.pyplot as plt # ডেটা ভিজুয়ালাইজেশনের জন্য ম্যাটপ্লটলিব ইম্পোর্ট।
from sklearn.datasets import load_iris # স্কিটলার্ন থেকে আইরিস ডেটাসেট লোড করার ফাংশন
```

```python
iris = load_iris()  # আইরিস ডেটাসেট লোড করে iris ভেরিয়েবলে রাখা হয়েছে।
```

```python
iris.feature_names # আইরিস ডেটাসেটের ফিচারের নামগুলো দেখাচ্ছে (যেমন sepal length, sepal width ইত্যাদি)।
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names) # আইরিসের ডেটাকে প্যান্ডাস ডেটাফ্রেমে রূপান্তর করা হচ্ছে।
df.sample(4) # ডেটাফ্রেম থেকে এলোমেলোভাবে ৪টি সারি দেখাচ্ছে।
```

```python
df['target']=iris.target # ডেটাফ্রেমে target নামে নতুন কলাম যোগ করা হচ্ছে।
df.sample(4)
```

```python
x = iris.data # ইনপুট ফিচার (ডেটা)
y = iris.target # আউটপুট টার্গেট (লেবেল)
```

```python
from sklearn.model_selection import train_test_split  #

x_train, x_test, y_train, y_test = train_test_split( # ডেটাকে ট্রেন ও টেস্ট সেটে ভাগ করা হচ্ছে:
    x, y,
    test_size = 0.2, # ২০% ডেটা টেস্ট, ৮০% ট্রেন।
    random_state = 42 # রিপ্রডিউসিবিলিটির জন্য সিড নির্ধারণ।
)
```

```python
from sklearn.tree import DecisionTreeClassifier # ডিসিশন ট্রি ক্লাসিফায়ার ইম্পোর্ট করা হচ্ছে।
```

```python
dtree = DecisionTreeClassifier() # dtree নামে ডিসিশন ট্রি অবজেক্ট তৈরি।
dtree = dtree.fit(x_train, y_train) #  ট্রেন ডেটা দিয়ে মডেল ট্রেন করা।
```

```python
y_pred = dtree.predict(x_test) # টেস্ট ডেটার উপর প্রেডিকশন করা এবং y_pred এ রাখা।
```

```python
from sklearn import tree
tree.plot_tree(dtree) # ডিসিশন ট্রি ভিজুয়ালাইজ করার জন্য plot_tree ব্যবহার করা হয়েছে। এটি ট্রির গঠন দেখায়।
```

```python
# টেস্টের আসল আউটপুট (y_test) এবং প্রেডিক্টেড আউটপুট (y_pred) দেখানো হচ্ছে।
y_test
y_pred
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test, y_pred) # from_predictions দিয়ে y_test এবং y_pred তুলনা করে কনফিউশন ম্যাট্রিক্স প্লট করা।

plt.show()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
# কনফিউশন ম্যাট্রিক্স, অ্যাকুরেসি, প্রিসিশন, রিকল, এফ১ স্কোর ক্যালকুলেট করা হচ্ছে।
cm = confusion_matrix(y_test, y_pred)  # সঠিক ও ভুল প্রেডিকশনের সংখ্যা দেখায় (TP, TN, FP, FN)।
accuracy = accuracy_score(y_test, y_pred) # অ্যাকুরেসি → (TP + TN) / (TP + TN + FP + FN) → মোট কতটা সঠিক।

# average='macro' → মাল্টিক্লাসের জন্য গড় মেট্রিক বের করতে।
precision = precision_score(y_test, y_pred, average='macro') # প্রিসিশন → TP / (TP + FP) → পজিটিভ বলেছিল যাদের, তাদের মধ্যে আসলে কতটা পজিটিভ
recall = recall_score(y_test, y_pred, average='macro') # রিকল (সেনসিটিভিটি) → TP / (TP + FN) → আসল পজিটিভের মধ্যে কতটা শনাক্ত করেছি।
f1 = f1_score(y_test, y_pred, average='macro') #  প্রিসিশন ও রিকলের হারমোনিক গড়:

# সব রেজাল্ট প্রিন্ট করা হয়েছে।
print("Confusion Matrix: \n", cm)
print("Accurayc: \n", accuracy)
print("Precision: \n", precision)
print("Recall: \n", recall)
print("F1: \n", f1)
```
